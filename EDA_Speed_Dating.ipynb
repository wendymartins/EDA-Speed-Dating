# 1 - Dataset Presentation

### Importing libraries and dataset

```python
import pandas as pd
import numpy as np 
import plotly.io as pio
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
from plotly.offline import plot
```

```python
# Displaying all columns of datasets
pd.set_option("display.max_columns", None)
pd.get_option("display.max_columns")

dataset = pd.read_csv("Speed Dating Data.csv", encoding = "ISO-8859-1")
dataset.head()
```

### Visualisations to better understand the sample

```python
print("Number of rows : {}".format(dataset.shape[0]))
print("Number of columns : {}".format(dataset.shape[1]))
print()

print("Basics statistics: ")
data_desc = dataset.describe(include='all')
display(data_desc)
print()

print("Percentage of missing values: ")
display(100*dataset.isnull().sum()/dataset.shape[0])
```

Many missing values, especially on the last columns. No need to look at them for further analysis.

There are also missing values on the "pid" column (=ID of the partner) : useless because it means the line does not correspond to a date.

We will drop them to make our analysis cleaner.

```python
dataset = dataset[~dataset["pid"].isna()]
dataset.shape
```

```python
print(f"There are {dataset['iid'].nunique()} unique iids in the dataset, which corresponds to the number of people studied.")
```

Some modifications on the dataset to facilitate its reading/understanding on graphs

```python
dataset['gender'] = dataset['gender'].apply(lambda x: 'Woman' if x == 0 else 'Man')
dataset['match'] = dataset['match'].apply(lambda x: 'Match' if x == 1 else 'No match')
dataset['samerace'] = dataset['samerace'].apply(lambda x: 'Same Race' if x == 1 else 'Different Race')
dataset['race'] = dataset['race'].apply(lambda x: 'Black' if x == 1 else 
                                                'European' if x == 2 else
                                                'Latino' if x == 3 else
                                                'Asian' if x == 4 else
                                                'Native American' if x == 5 else 'Other')
dataset['imprace'] = dataset['imprace'].apply(lambda x: 'Not important' if x <= 3 else 
                                                'Moderately important' if x <=  7 else 'Important')

dataset['imprelig'] = dataset['imprelig'].apply(lambda x: 'Not important' if x <= 3 else 
                                                'Moderately important' if x <=  7 else 'Important')
```

New columns to analyse some data afterwards.

```python
dataset['Age_diff'] = abs(dataset['age'] - dataset['age_o'])
```

Creating a new dataset : dropping duplicates on iid to be able to show statistics on a personal basis
*(example : gender, age, race etc (everything related to the person/iid is always the same, no matter who he/she meets)*.

```python
dataset_unique_iid = dataset.drop_duplicates("iid")
dataset_unique_iid.head()
```

Proportion of Men/Women

```python
fig = px.pie(dataset_unique_iid, names="gender", title="Gender Ratio", hole=0.4,
            color=dataset_unique_iid["gender"],
            color_discrete_map={'Man':'blue',
            'Woman':'lightpink'})
fig.update_layout(title_x=0.48)
fig.show()
```

Age representation

```python
fig = px.violin(dataset_unique_iid, y="gender", x="age", color="gender", box=True, points="all",
            color_discrete_map={'Man':'blue','Woman':'lightpink'}, title="Age representation")
fig.show()
```

```python
mean_age = dataset_unique_iid["age"].mean()
std_age = dataset_unique_iid["age"].std()

print(f"If we look at the dispersion around the mean, most of the participants are aged between {round(mean_age - std_age)} and {round(mean_age + std_age)}.")
```

In a normal distribution, we have 66% of data between **the mean - 1 standard deviation** and **the mean + 1 standard deviation**.

Even if we can almost guess it from the violin plot, let's check the curve distribution.

```python
age_w = dataset_unique_iid[dataset_unique_iid["gender"] == "Woman"]["age"]
age_w = age_w[age_w.notnull()]

age_m = dataset_unique_iid[dataset_unique_iid["gender"] == "Man"]["age"]
age_m = age_m[age_m.notnull()]
```

```python
hist_data = [age_m]
group_labels = ['Men curve']
colors=['blue']

fig = ff.create_distplot(hist_data, group_labels, colors=colors)
fig2 = ff.create_distplot(hist_data, group_labels, curve_type='normal')

normal_x = fig2.data[1]['x']
normal_y = fig2.data[1]['y']

fig.add_traces(go.Scatter(x=normal_x, y=normal_y, mode = 'lines',
                          line = dict(color='green', 
                                      width = 2),
                          name = 'Normal curve'
                         ))
fig.update_layout(xaxis_range=[17,60])
fig.update_layout(title="Men Age Distribution", title_x=0.4)
fig.show()
```

Since distribution is not exactly normal, it is better to check the dispersion around the median.

We can see it does not change much.

```python
Q1_age = dataset_unique_iid["age"].quantile(0.25)
Q3_age = dataset_unique_iid["age"].quantile(0.75)

print(f"If we look at the dispersion around the median (which could be more accurate in this case since we have outliers),\n we know 50% of the participants are aged between {Q1_age} and {Q3_age}.")
```

> Our data:
>
> Study: experimental *(subjects put on a specific situation)*, longitudinal *(subjects are being followed up over time)*, stratified *(almost as many women as men).
>
> Data: quantitative disecrete *(age, ratings), qualitative nominal *(gender, origin)

### Assumption: Similar people tend to associate to each other

There is a French saying that goes: "Qui se ressemble s'assemble". It means that similar people tend to associate to each other.

We will try to figure out if we can validate it with our data.

# 2 - Checking Assumptions

```python
colors = ['red','', 'green']

fig = go.Figure(data=[go.Pie(labels=dataset["match"], pull=[0, 0, 0.2, 0])])
fig.update_layout(title_text="Matching Ratio", title_x = 0.47)
fig.update_traces(marker=dict(colors=colors))
fig.show()
```

### Is age important for matching?

```python
fig = px.histogram(dataset, x="Age_diff", color='match', barmode='group', title="Age difference during dates",
                    color_discrete_map = {"Match":'green', "No match":'red'})
fig.update_layout(bargap=0.2, xaxis_title="Age difference", legend_title_text="")
fig.show()
```

> The proportion of match seems to stay the same, no matter what the age difference is.

### Is origin important for matching?

Let's see what happened within same origin dates and different origin ones.

```python
fig = px.sunburst(dataset, path=['samerace', 'match'], title="Percentage of match within same or different origin dates",
                color='match', color_discrete_map={'(?)':'lightgrey','Match':'green', 'No match':'red'})
fig.update_layout(title_x = 0.5)
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

Question of importance was asked to participants.

```python
df_women_all= dataset[dataset["gender"] == "Woman"]
df_men_all= dataset[dataset["gender"] == "Man"]

colors=('#FF8900','#FFB866',"#FFDBB0")

fig = make_subplots(rows=1, cols=2, specs=[[{"type": "pie"}, {"type": "pie"}]])

fig.add_trace(go.Pie(labels=df_women_all["imprace"], hole = 0.6),row=1, col=1)

fig.add_trace(go.Pie(labels=df_men_all["imprace"], hole = 0.6,marker_colors=(colors)),row=1, col=2)

fig.update_layout(showlegend=True,title_text="Importance in getting a partner with same origin for Women and Men", title_x=0.43,
                annotations=[dict(text='Women', x=0.19, y=0.5, font_size=20, showarrow=False),
                                dict(text='Men', x=0.8, y=0.5, font_size=20, showarrow=False)])

fig.show()
```

Now let's look only at dates that matched.

We can see if people who think that it is important to have the same origin only matched with people from the same origin.

We can also check if people who say it is not important get more matches with people from the same or different origin.

```python
df_women_match= dataset[(dataset["gender"] == "Woman") & (dataset["match"] == "Match")]

fig = px.sunburst(df_women_match, path=['imprace', 'samerace'],title="Proportion of match with origin information related to importance given by Women",
                color='samerace', color_discrete_map={'(?)':'lightgrey','Same Race':'#FF8900', 'Different Race':'FFDBB0'})
fig.update_layout(title_x = 0.5)
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

```python
df_men_match= dataset[(dataset["gender"] == "Man") & (dataset["match"] == "Match")]

fig = px.sunburst(df_men_match, path=['imprace', 'samerace'],title="Proportion of match with origin information related to importance given by Men",
                color='samerace', color_discrete_map={'(?)':'lightgrey','Same Race':'#FF8900', 'Different Race':'FFDBB0'})
fig.update_layout(title_x = 0.5)
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

> As numbers show, having the same origin is not an important factor, neither in what people say nor on the facts observed.

# 3 - The need to know each other to decide what is important

Trying to analyse better the differences between what people want, what they think others want, and what they think of themselves.

### What people think the other sex is looking for VS what the other sex is actually looking for

```python
# What women think men are looking for, compared to what men are really looking for.

mean_att_others_by_women = df_unique_woman["attr2_1"].mean()
mean_sinc_others_by_women = df_unique_woman["sinc2_1"].mean()
mean_inte_others_by_women = df_unique_woman["intel2_1"].mean()
mean_fun_others_by_women = df_unique_woman["fun2_1"].mean()
mean_ambi_others_by_women = df_unique_woman["amb2_1"].mean()
mean_shar_others_by_women = df_unique_woman["shar2_1"].mean()

mean_att_men = df_unique_man["attr1_1"].mean()
mean_sinc_men = df_unique_man["sinc1_1"].mean()
mean_inte_men = df_unique_man["intel1_1"].mean()
mean_fun_men = df_unique_man["fun1_1"].mean()
mean_ambi_men = df_unique_man["amb1_1"].mean()
mean_shar_men = df_unique_man["shar1_1"].mean()

categories = ['attractive', 'sincere','intelligent','fun', 'ambitious', 'shared interests']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[mean_att_men,mean_sinc_men, mean_inte_men, mean_fun_men, mean_ambi_men, mean_shar_men],
      theta=categories,
      fill='toself',
      name='What men want'
))

fig.add_trace(go.Scatterpolar(
      r=[mean_att_others_by_women,mean_sinc_others_by_women, mean_inte_others_by_women, mean_fun_others_by_women,
      mean_ambi_others_by_women,mean_shar_others_by_women],
      theta=categories,
      fill='toself',
      name='What women think men want'
))

fig.update_layout(polar=dict(radialaxis=dict(visible=True,range=[0, 40])),showlegend=True,
                  title_text="What women think men want VS what men really want", title_x=0.42)

fig.show()
```

```python
# What men think women are looking for compared to what women are really looking for.

mean_att_others_by_men = df_unique_man["attr2_1"].mean()
mean_sinc_others_by_men = df_unique_man["sinc2_1"].mean()
mean_inte_others_by_men = df_unique_man["intel2_1"].mean()
mean_fun_others_by_men = df_unique_man["fun2_1"].mean()
mean_ambi_others_by_men = df_unique_man["amb2_1"].mean()
mean_shar_others_by_men = df_unique_man["shar2_1"].mean()

mean_att_women = df_unique_woman["attr1_1"].mean()
mean_sinc_women = df_unique_woman["sinc1_1"].mean()
mean_inte_women = df_unique_woman["intel1_1"].mean()
mean_fun_women = df_unique_woman["fun1_1"].mean()
mean_ambi_women = df_unique_woman["amb1_1"].mean()
mean_shar_women = df_unique_woman["shar1_1"].mean()

categories = ['attractive', 'sincere','intelligent','fun', 'ambitious', 'shared interests']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[mean_att_others_by_men,mean_sinc_others_by_men, mean_inte_others_by_men, mean_fun_others_by_men,
      mean_ambi_others_by_men,mean_shar_others_by_men],
      theta=categories,
      fill='toself',
      name='What men think women want'
))

fig.add_trace(go.Scatterpolar(
      r=[mean_att_women,mean_sinc_women, mean_inte_women, mean_fun_women, mean_ambi_women, mean_shar_women],
      theta=categories,
      fill='toself',
      name='What women want'
))

fig.update_layout(polar=dict(radialaxis=dict(visible=True,range=[0, 40])),showlegend=True,
                  title_text="What men think women want VS what women really want", title_x=0.42)

fig.show()
```

> In general, we can say that people seem to know what the opposite sex is looking for.
>
> Even if for both men and women, there is an overestimation about the attractiveness the other sex is looking for.
>
> And in the same way, there is an underestimation about the intelligence and sincerity the other sex is looking for.

### Are people looking for someone like them?

```python
# Comparison between what women want and what they think of themselves.

mean_att_self_women = df_unique_woman["attr3_1"].mean()
mean_sinc_self_women = df_unique_woman["sinc3_1"].mean()
mean_inte_self_women = df_unique_woman["intel3_1"].mean()
mean_fun_self_women = df_unique_woman["fun3_1"].mean()
mean_ambi_self_women = df_unique_woman["amb3_1"].mean()

fig = make_subplots(rows=1, cols=2, specs=[[{'type': 'polar'}] * 2] * 1)

fig.add_trace(go.Scatterpolar(
      r=[mean_att_women,mean_sinc_women, mean_inte_women, mean_fun_women, mean_ambi_women],
      theta=categories,
      fill='toself',
      name='What women want'
), row=1, col=1)


fig.add_trace(go.Scatterpolar(
      r=[mean_att_self_women,mean_sinc_self_women, mean_inte_self_women, mean_fun_self_women, mean_ambi_self_women],
      theta=categories,
      fill='toself',
      name='Women opinion about themselves'
), row=1, col=2)


fig.update_layout(polar1=dict(radialaxis=dict(visible=False)),polar2=dict(radialaxis=dict(visible=False)),showlegend=True,
                  title_text="Is what women are looking for the same as what they think of themselves?", title_x=0.42)

fig.show()
```

```python
# Comparison between what men want and what they think of themselves.

mean_att_self_men = df_unique_man["attr3_1"].mean()
mean_sinc_self_men = df_unique_man["sinc3_1"].mean()
mean_inte_self_men = df_unique_man["intel3_1"].mean()
mean_fun_self_men = df_unique_man["fun3_1"].mean()
mean_ambi_self_men = df_unique_man["amb3_1"].mean()

fig = make_subplots(rows=1, cols=2, specs=[[{'type': 'polar'}] * 2] * 1)

fig.add_trace(go.Scatterpolar(
      r=[mean_att_men,mean_sinc_men, mean_inte_men, mean_fun_men, mean_ambi_men],
      theta=categories,
      fill='toself',
      name='What men want'
), row=1, col=1)


fig.add_trace(go.Scatterpolar(
      r=[mean_att_self_men,mean_sinc_self_men, mean_inte_self_men, mean_fun_self_men, mean_ambi_self_men],
      theta=categories,
      fill='toself',
      name='Men opinion about themselves'
), row=1, col=2)


fig.update_layout(polar1=dict(radialaxis=dict(visible=False)),polar2=dict(radialaxis=dict(visible=False)),showlegend=True,
                  title_text="Is what men are looking for the same as what they think of themselves?", title_x=0.42)

fig.show()
```

> People do not seem interested in having a partner that is like them.
> 
> For example, men think they are almost as attractive, as they are intelligent, fun, sincere and ambitious. But they are looking for women that are especially attractive. Others qualities having less importance.
> 
> This is a bit less pronounced for women.

# Conclusion

There are similarities between men and women.

People are not looking for someone like them (example of what people are looking for). But they are not looking for someone completely different either (example of activities).

These conclusions are made knowing that we have bias: the sample is not representative of the real population, people here are put in a specific situation, and they know they are being observed.
Additionnaly, all people were not exactly in the same position: some met 5 people, others met more than 20; some people met in winter and others before summer. These dates were only within different gender dates, not taking into account gay people.
